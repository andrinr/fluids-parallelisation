```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import  LogNorm
from scipy.io import FortranFile
```

```python

with FortranFile('exact_solution_00000', 'r') as f:
    nx1, ny1 = f.read_ints('i')
    exact1 = f.read_reals('f4')

with FortranFile('exact_solution_00001', 'r') as f:
    nx2, ny2 = f.read_ints('i')
    exact2 = f.read_reals('f4')
```

```python
exact1 = np.array(exact1)
exact1 = exact1.reshape(ny1, nx1)
exact2 = np.array(exact2)
exact2 = exact2.reshape(ny2, nx2)
```

```python
exact = np.concatenate((exact1,exact2), axis=1)
```

```python
plt.imshow(exact)
plt.colorbar()
```
